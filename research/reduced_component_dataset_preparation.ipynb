```python
import sys
sys.path.append('../..')

import os
import json
import ipywidgets as widgets

from source.settings import DATASETS_FOLDER
```

```python
DATASET_PATH = f'{DATASETS_FOLDER}/fpic-component-inpainted'

lbl = widgets.Label()
status = widgets.Label()
im_dot = widgets.Image(format='png', width=300, height=300)
im_inp0 = widgets.Image(format='png', width=300, height=300)
im_inp1 = widgets.Image(format='png', width=300, height=300)
im_inp2 = widgets.Image(format='png', width=300, height=300)

# index = {}
# with open('./_index.json', 'r') as f:
#     index = json.load(f)

def walk():
    _ = yield
    for root, _, files in os.walk(DATASET_PATH):
        files = [f for f in files if 'index' not in f]
        if len(files) and root not in index:
            lbl.value = root
            im_dot_file = open(f"{root}/with_points.png", "rb")

            im_inp0_file = open(f"{root}/inpainted_with_mask_0.png", "rb")
            im_inp1_file = open(f"{root}/inpainted_with_mask_1.png", "rb")
            im_inp2_file = open(f"{root}/inpainted_with_mask_2.png", "rb")

            im_dot.value = im_dot_file.read()
            im_inp0.value = im_inp0_file.read()
            im_inp1.value = im_inp1_file.read()
            im_inp2.value = im_inp2_file.read()

            im_dot_file.close()
            im_inp0_file.close()
            im_inp1_file.close()
            im_inp2_file.close()

            status.value = ''

            yield None


def push():
    with open('./_index.json', 'w') as f:
        json.dump(index, f)

    status.value = status.value + ' [ok]'


def on_next(*args, **kwargs):
    next(walker)


def on_b0(*args, **kwargs):
    index[lbl.value] = 'inpainted_with_mask_0.png'
    status.value = '--> 0'
    push()


def on_b1(*args, **kwargs):
    index[lbl.value] = 'inpainted_with_mask_1.png'
    status.value = '--> 1'
    push()


def on_b2(*args, **kwargs):
    index[lbl.value] = 'inpainted_with_mask_2.png'
    status.value = '--> 2'
    push()


walker = walk()
next(walker)
next(walker)

b_next = widgets.Button(description='Next')
b_next.on_click(on_next)

b_0 = widgets.Button(description='0')
b_0.on_click(on_b0)
b_1 = widgets.Button(description='1')
b_1.on_click(on_b1)
b_2 = widgets.Button(description='2')
b_2.on_click(on_b2)

widgets.VBox([
    widgets.HBox([lbl, status]),
    im_dot,
    widgets.HBox([im_inp0, im_inp1, im_inp2]),
    widgets.HBox([b_0, b_1, b_2, b_next])
])
```

```python
im_org = widgets.Image(format='png', width=300, height=300)
im_inp = widgets.Image(format='png', width=300, height=300)
lbl = widgets.Label()

# with open('./_index.json', 'r') as f:
#     index = json.load(f)
# with open('./_index_clean.json', 'w') as f:
#     json.dump(index, f)


def view():
    _ = yield
    for root, file in index.items():
        lbl.value = root
        org_file = open(f"{root}/with_points.png", "rb")
        inp_file = open(f"{root}/{file}", "rb")

        im_org.value = org_file.read()
        im_inp.value = inp_file.read()

        org_file.close()
        inp_file.close()
        yield None


def on_next(*args, **kwargs):
    next(viewer)


def on_del(*args, **kwargs):
    with open('./_index_clean.json', 'r') as f:
        tmp = json.load(f)
    if lbl.value in tmp:
        del tmp[lbl.value]
        with open('./_index_clean.json', 'w') as f:
            json.dump(index, f)



viewer = view()
next(viewer)
next(viewer)

b_next = widgets.Button(description='Next')
b_next.on_click(on_next)

b_del = widgets.Button(description='Delete')
b_del.on_click(on_del)

widgets.VBox([
    widgets.HBox([lbl]),
    widgets.HBox([im_org, im_inp]),
    widgets.HBox([b_del, b_next])
])
```
